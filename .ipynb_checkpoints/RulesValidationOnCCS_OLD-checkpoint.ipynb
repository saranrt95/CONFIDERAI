```python
import pandas as pd
import numpy as np
import re
```

```python
def evaluate_rule_conditions(row, condition_part):

    # Checks if any of the conditions in the rule are satisfied
    if all(check_condition(row, part) for part in condition_part.split(" and ")):
        return True  # Return True if all conditions in the rule are satisfied
    
    return False  # Return False if any of the conditions in the rule is not satisfied


def check_condition(row, condition_part):
    # Check if a single condition part is satisfied
    
    #parts = [part.strip('()') for part in condition_part.split()]
    #parts = [part for part in condition_part.split()]
    # Use regular expressions to properly parse the condition
    parts = re.split(r'\s*(==|<=|>=|<|>|!=)\s*', condition_part)
    
    #print("parts: ", parts)
    if len(parts) == 3:
        column, op, value = parts
        return eval(f"{row[column]} {op} {value}")
    # handle the case of a 2-thresholds conditions of the kind: a < Column <= b
    elif len(parts) == 5:
        val1,op1,column,op2,val2 = parts
        # Use the original condition from the rule
        return eval(f"{val1} {op1} {row[column]} {op2} {val2}")        
    else:
        RaiseValueError("Bad condition formatting!")


def evaluate_rules(data, N1, tuned_rules, output, dataset_name = None):

    idx_rules = 0
    satisfiedMat = np.zeros((len(data),len(tuned_rules)))
    coveringlist = []
    errorlist = []
    precisionlist = []
    for i, rule in tuned_rules.iterrows():
        pred_rule = []
        tptot = 0
        tntot = 0
        fntot = 0
        fptot = 0
        
        idx_data = 0
        tuned_antecedent = rule['Rule'].strip()
        #print(f"rule: {tuned_antecedent}")
        for j, row in data.iterrows():
            y_row = row[output]
            tp = 0
            tn = 0
            fn = 0
            fp = 0
            # check if the point row satifies rule 
            if evaluate_rule_conditions(row, tuned_antecedent):
                # rule is satisfied
                satisfied = True              
                satisfiedMat[idx_data,idx_rules] = 1
            else:
                satisfied = False
    
            
            if satisfied and y_row == 1:
                tp=1
            elif satisfied and y_row == 0:
                fp=1
            elif not satisfied and y_row == 0:
                tn=1
            
            elif not satisfied and y_row == 1:
                fn=1

            tptot+=tp
            tntot+=tn
            fptot+=fp
            fntot+=fn
            idx_data+=1
        if (tptot+fptot) == 0 and (fptot+tntot)==0: 
            coveringlist.append(tptot/(tptot+fntot))
            errorlist.append(0)
            precisionlist.append(0)
            continue
        elif (tptot+fptot) == 0:
            err = fptot/(tntot+fptot)
            coveringlist.append(tptot/(tptot+fntot))
            errorlist.append(err)
            precisionlist.append(0)
            continue
        elif (tntot+fptot)==0:
            errorlist.append(0)
            cov = tptot/(tptot+fntot)
            prec = tptot/(tptot+fptot)
            coveringlist.append(cov)
            precisionlist.append(prec)
            continue
        elif (tptot+fntot)==0:
            coveringlist.append(0)
            err = fptot/(tntot+fptot)
            prec = tptot/(tptot+fptot)
            errorlist.append(err)
            precisionlist.append(prec)
            continue
        else:
            #print(f"tp = {tptot}, fp = {fptot}, tn = {tntot}, fn = {fntot}")
            #cov = tptot/N1
            cov = tptot/(tptot+fntot)
            err = fptot/(tntot+fptot)
            prec = tptot/(tptot+fptot)
            coveringlist.append(cov)
            errorlist.append(err)
            precisionlist.append(prec)
            continue
        idx_rules+=1
    
    return satisfiedMat, coveringlist, precisionlist, errorlist


```

```python
datasetlist=['ssh', 'p2p', 'smoking', 'cardio', 'platooning', 'rul', 'eeg', 'mqttset', 'telescope', 'fire']
outputlabels = ['g','g', 'smoking', 'cardio','collision','RUL_binary','eyeDetection','output','class','Fire Alarm']
#TEST_CCS = False

validation_ccs = []

for dataset_name,output_label in list(zip(datasetlist,outputlabels)):
    # Loading rules
    rules = pd.read_csv(f"/Users/saranarteni/OneDrive - CNR/Conformal Predictors/Dataset20230619/{dataset_name}/{dataset_name}_rules.csv", header=None, names=["Rule", "Covering", "Error"])
    
    # adjust columns values
    for i in range(len(rules)+1):   
        rules["Rule"] = rules["Rule"].apply(lambda x: x.replace("RULE {}: ".format(i),""))
    rules["Rule"] = rules["Rule"].apply(lambda x: x.replace("AND","and"))
    rules["Rule"] = rules["Rule"].apply(lambda x: x.replace("{",""))
    rules["Rule"] = rules["Rule"].apply(lambda x: x.replace("}",""))
    rules["Rule"] = rules["Rule"].apply(lambda x: x.replace(f"{output_label} in ",f"{output_label} = "))
    rules["Covering"] = rules["Covering"].apply(lambda x: x.replace("COVERING: ",""))
    rules["Error"] = rules["Error"].apply(lambda x: x.replace("ERROR: ",""))
    rules['Output'] = rules['Rule'].str.extract(rf'{output_label} = (\d+)', expand=False).astype(int)
    rules["Rule"] = rules["Rule"].apply(lambda x: x.replace("IF ",""))
    rules["Rule"] = rules["Rule"].apply(lambda x: x.replace(x[x.find("THEN"):],""))
    
    # load test data for epsilon 0.05
    data = pd.read_excel(f"results_rev_paper_20240325/results_eps005_{dataset_name}.xlsx")
    dataCCS = data[data["ConformalPoint"] == 1]
    #dataORIG = data[data[output_label] == 1]
    dataORIG = data
    N1 = len(data[data[output_label] == 1])
    rules_targetcls = rules[rules["Output"]==1]
    _, coveringvaluesCCS,precisionCCS,errorCCS = evaluate_rules(dataCCS, N1, rules_targetcls, output_label,dataset_name = dataset_name)
    _, coveringvaluesORIG,precisionORIG,errorORIG = evaluate_rules(dataORIG, N1, rules_targetcls, output_label,dataset_name = dataset_name)
    df = pd.DataFrame(data = zip([dataset_name]*len(rules_targetcls),list(rules_targetcls["Rule"]),coveringvaluesCCS,coveringvaluesORIG,precisionCCS,precisionORIG,errorCCS,errorORIG), columns = ["Dataset","Rule","CoveringCCS","CoveringOriginal","PrecisionCCS","PrecisionOriginal","ErrorCCS","ErrorOriginal"])
    validation_ccs.append(df)

ValDf = pd.concat(validation_ccs, ignore_index = True)
```

# Metrics analysis


```python
metrics = pd.read_excel("results_rev_paper_20240325/validationCCS_v5.xlsx")
metrics
```

```python
datasetlist=['ssh', 'p2p', 'smoking', 'cardio', 'platooning', 'rul', 'eeg', 'mqttset', 'telescope', 'fire']

ssh = metrics[metrics["Dataset"]=="ssh"]
ssh
```

```python
variations_ccs = []

for d in datasetlist:
    db = metrics[metrics["Dataset"]==d]  
    var_rel = list(((db["RelevanceCCS"]-db["RelevanceOriginal"])/db["RelevanceOriginal"])*100)
    var_prec = list(((db["PrecisionCCS"]-db["PrecisionOriginal"])/db["PrecisionOriginal"])*100)
    variations_ccs.append(pd.DataFrame(data = zip([d]*len(db),var_rel,var_prec), columns = ["Dataset", "RelevanceVariation", "PrecisionVariation"]))
```

```python
variations_percentage = pd.concat(variations_ccs, ignore_index = True)
variations_percentage = variations_percentage.fillna(value = 0)
variations_percentage
```

```python
variations_percentage.describe()
```

```python
import matplotlib.pyplot as plt

#d = "ssh"
for d in datasetlist:
    db = variations_percentage[variations_percentage["Dataset"]==d]  
    rules_idx = [i+1 for i in list(range(len(db)))]
    fig, ax1 = plt.subplots(figsize=(15, 6))
    ax1.scatter(rules_idx, db["RelevanceVariation"], marker='o', color='b', label='Relevance')
    ax1.set_xlabel('Rule', fontsize = 24)
    ax1.set_ylabel('$\Delta_{\%}$ Relevance', fontsize = 24)
    ax1.tick_params(axis='y', labelcolor='k', labelsize = 18)
    ax1.set_xticks(rules_idx)
    ax1.set_xticklabels(rules_idx, fontsize = 20, rotation = 45)
    
    # Abilitazione della griglia
    ax1.grid(True)
    
    ax2 = ax1.twinx()
    ax2.scatter(rules_idx, db["PrecisionVariation"], marker='x', color='r', label='Precision')
    ax2.set_ylabel('$\Delta_{\%}$ Precision', fontsize = 24)
    ax2.tick_params(axis='y', labelcolor='k', labelsize = 18)
    
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc='center right', fontsize = 16)
    #plt.title(f"{d.upper()}", fontsize = 24)
    #plt.xticks(rotation=45) 
    plt.savefig("results_rev_paper_20240325/variations_"+d+".png", dpi = 200)
    plt.tight_layout()
    plt.show()
```
